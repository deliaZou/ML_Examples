## 1. 时间序列基础

时间序列是在一系列时间点上获取的的数据构成的序列。时间序列数据相对于普通数据的区别在于时间序列数据包含顺序信息。时间序列分析的目的是通过掌握数据变化的规律对未来数据进行预测。时间序列主要的研究对象是序列上的数据与其自身滞后项及误差项之间的关系。

对时间序列建模的前提假设：
* 时间序列的数值与其滞后项之间存在关联关系
* 时间序列满足宽平稳条件（即统计特征不随时间改变）

### 1.2. 时间序列平稳性

#### 1.2.1. 平稳性

什么是平稳：
* 严平稳：任意时间间隔下，时间序列变量的联合分布不随时间改变。
* 宽平稳：随机过程的均值和方差都为常函数，且任意时间间隔随机变量与其滞后项的协方差只与时间间隔有关，与具体的时间点无关

由于严平稳条件难以验证，一般分析过程中只会考虑宽平稳的条件。

#### 1.2.2. 单位根检验

平稳性检验的两种常用方法：
* 图形检验法：平稳时间序列不应该包含趋势性、季节性或者其他依赖于时间的特性
* 描述性统计结果：将时间序列分组，如果各组的均值方差等统计特征有显著差异，时间序列通常是非平稳的
* 统计假设检验：单位根检验法

滞后算子（Backshift Operator）：时间序列中经常用 $B$ 表示滞后算子：

$By_t = y_{t-1}$

滞后算子有如下类似代数运算的性质：

$B(By_{t}) = B^{2}y_{t} = y_{t-2}$

如果用 $y_t'$ 表示 $y_t$ 的一阶差分，$y_t''$ 表示 $y_t$ 的二阶差分，则有：

$y'_{t} = y_{t} - y_{t-1} = y_t - By_{t} = (1 - B)y_{t}$

$y_t'' = y_{t} - 2y_{t - 1} + y_{t - 2} = (1-2B+B^2)y_t = (1 - B)^{2} y_{t}$

一个时间序列可以使用滞后算子表示，比如时间序列 $y_t = a_1y_{t-1} - a_2y_{t-2} + w_t$ 可以表示为 $(1-a_1B + a_2B^2)y_t = w_t$ 。时间序列平稳的条件是滞后算子多项式（此处是 $1-a_1B + a_2B^2 =0$）的所有根的绝对值都大于1，滞后算子多项式如果有为1的根，则称为时间序列存在单位根。存在单位根的时间序列为非平稳时间序列。时间序列的这种特性可以通过简单的一阶自回归过程 $y_t = \alpha y_{t-1} + w_t$观察到。

```python
import warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
from pylab import mpl
from statsmodels.tsa.stattools import adfuller 

warnings.filterwarnings(action='once')
%matplotlib inline
sns.set_style('darkgrid')
mpl.rcParams['figure.figsize'] = (14, 4)
```

```python
# 定义函数模拟一个一阶自回归过程
def generate_ts(alpha=1, steps=300):
    w = np.random.randn(steps)
    ts = [0]
    for i in range(1, len(w)):
        ts.append(alpha * ts[i-1] + w[i])
    return ts
```

```python
# 模拟一个随机游走过程，即alpha=1
ts = generate_ts(1)
plt.plot(ts)
```

```python
# 比较不同的alpha产生的时间序列
alpha = [0.8, 1, 1.01]
ts_df = pd.DataFrame([generate_ts(a) for a in alpha], index=alpha).T
ts_df.plot(subplots=True, figsize=(14, 9))
```

通过比较不同 $\alpha$ 值产生的时间序列图形可以看到，当特征根绝对值小于1时，时间序列是平稳的（方差不会趋于无穷），当特征根的绝对值大于等于1时，时间序列的方差会随着时间的增长趋于无穷。所以统计学里通过单位根检验判别时间序列的平稳性。单位根检验的思路是拟合一个 $y_t = \alpha_0 + \alpha_1y_{t-1} + w_t$ 过程，判别 $\alpha_1$ 与 1 是否有显著差异，即判别时间序列是否存在单位根。

单位根检验是一种假设检验的方法：
* 原假设：时间序列存在单位根，即时间序列非平稳
* 备择假设：时间序列不存在单位根，即时间序列平稳

对单位根检验结果的解读：
* 如果p值小于等于约定的阈值（一般是1%或5%），则拒绝原假设，时间序列不存在单位根，因此是平稳序列
* 否则不能拒绝原假设，不能判定时间序列为平稳序列

进行单位根检验，可以使用statsmodels提供的adfuller函数。statsmodels文档：http://www.statsmodels.org/dev/generated/statsmodels.tsa.stattools.adfuller.html

注：导入statsmodel过程如果出现报错，可以通过 pip install --upgrade statsmodels 升级工具包。升级完成后重启notebook内核一般可以解决。

```python
# 读入原始数据
data = pd.read_csv('data/600240.csv', parse_dates=True, index_col='date')
data['d_return'] = data['d_return'] * 100
data.head()
```

```python
# 对原始数据进行切割，将1/4的数据划分为测试数据
# train, test = data[:480], data[480:]
# len(train)
```

```python
# 观察数据特征
data.plot(subplots=True, figsize=(10, 6))
```

```python
def adf_test(s):
    """对序列s进行ADF检验，打印检验结果"""
    return_adf = adfuller(s)
    print(f'adf: {return_adf[0]: .2f},\n'
    f'p-value: {return_adf[1]: .2f},\n'
    f'critical values:\n'
    f'1%：{return_adf[4]["1%"]: .2f}\n'
    f'5%：{return_adf[4]["5%"]: .2f}\n'
    f'10%：{return_adf[4]["10%"]: .2f}\n')
```

```python
adf_test(data.d_return)
```

可以看到，收益率数据的adf统计值远小于1%显著度对平对应的阈值，所以p-value远远小于1%，可以否定时间序列存在单位根的原假设，即测试的序列是平稳的。

```python
# 进行单位根检验
adf_test(data.close)
```

收盘价数据adf检测的p-value大于5%，因此不能否定原假设，即时间序列非平稳。

#### 1.2.3. 非平稳序列转化为平稳序列

对于非平稳的时间序列，一般首先要将其转化为平稳序列（统计特征不随时间变化），再对其进行建模。

造成时间序列非平稳的原因一般是存在趋势或周期因素，对非平稳时间序列的处理一般有以下几种方法：
* 对原始数据进行差分操作，对于趋势性和周期性因素都有效果
* 对原始时间序列取对数或平方根，一般使用于处理存在趋势的时间序列
* 首先对原始数据的趋势性和周期性建模，之后用平稳模型拟合残差，即认为残差是平稳的

```python
# 对价格数据进行差分操作
close_diff = data.close.diff().dropna()
close_diff.head()
```

```python
# 观察差分以后的结果
close_diff.plot()
```

```python
# 对差分结果进行ADF检验
adf_test(close_diff)
```

可以看到，价格数据经过差分以后，ADF检验的p-value已经小于1%和5%的显著度水平，即差分后的序列已经是平稳序列。

### 1.3. 时间序列自相关性

相关性研究的是两个变量之间的关系，自相关性研究的是时间序列上的数据与其自身的滞后项之间的关系。时间序列分析中，可以通过自相关性判断序列的随机性，并对非随机的序列选择适合的模型。时间序列的自相关性一般用自相关系数函数（ACF）和偏自相关系数函数（PACF）衡量。

#### 1.3.1. 自相关系数函数（Autocorrelation Coefficient Function，ACF）

对于平稳时间序列，自相关系数方程可以表示为：


$ r_k = \displaystyle \frac{Cov(X_i, X_{i-k})}{Var(X_i)} = \frac{E[(X_i - \mu_X)(X_{i-k} - \mu_X)]}{E [(X_i - \mu_X)^2]}$


其中 $Cov(X_i, X_{i-k})$ 是时间序列的自协方差，$r_k$ 可以看做是关于时间间隔$k$的函数。

将k从0到n取值计算得到的系相关系数画在同一张图上就得到自相关图。

自相关系数描述了时间序列和其自身滞后项之间的相关性，这种相关性同时包含了直接相关和间接相关。

#### 1.3.2. 偏自相关系数函数（Partial Autocorrleation Coefficient Function, PACF）

如果时间序列 $X_i$ 与时间序列 $X_{i-k}$ 存在相关关系，即 $X_i$ 受到 $X_{i-k}$ 的影响，则 $X_{i-k}$ 也会受到 $X_{i-2k}$ 的影响，因此 $X_{i-2k}$ 会间接地通过 $X_{i-k}$ 影响到 $X_i$。偏自相关系数消除了所有低阶滞后项的间接相关性影响，只反映时间序列与其自身滞后项之间的直接相关关系。也可以说偏自相关系数反映了时间序列与其滞后项之间没有被其他低阶滞后项解释的相关性。

#### 1.3.3. 自相关和偏自相关图的使用

计算时间序列的自相关系数和偏自相关系数，可以使用statsmodels工具包中的acf()函数和pacf()函数。实际中自相关和偏自相关的图形更常使用。

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# 不设置最大滞后项的自相关系数函数图
plot_acf(data.close, use_vlines=True, title='close acf')
plt.show()
```

不设置最大滞后项的情况下，自相关系数图会显示所有阶数的自相关系数。自相关系数函数是滞后时间与相关系数之间的函数。一般滞后时间越长，相关系数越趋近于0。可以理解为间隔的时间越长，过去的数据与现在的数据相关性越小；或以现有数据为基础，对未来进行预测，预测的时间约长难度越大。

```python
# 自相关图和偏自相关图
fig, ax = plt.subplots(2, 1, figsize=(10, 6))
plot_acf(data.close, use_vlines=True, lags=30, ax=ax[0], title='close acf')
plot_pacf(data.close, use_vlines=True, lags=30, ax=ax[1], title='close pacf')
plt.show()
```

```python
# 定义函数绘制自相关和偏自相关图
def plot_autoregression(s, title="", lags=30):
    fig, ax = plt.subplots(2, 1, figsize=(10, 6))
    plot_acf(s, use_vlines=True, lags=lags, ax=ax[0], title=f'{title} acf')
    plot_pacf(s, use_vlines=True, lags=lags, ax=ax[1], title=f'{title} pacf')
    plt.show()
```

从自相关图上可以看到，收盘价数据与高阶滞后项之间仍然存在很大的相关性。但从偏自相关图上看，只有一阶偏自相关系数明显大于其他滞后项。说明高阶滞后项的相关性都可以用一阶滞后项的相关性解释。

关于自相关图的两个常用描述：
* 截尾：相关系数在k个滞后项以后快速趋近于0，即k阶截尾
* 拖尾：相关系数始终有大于0的取值

图中阴影部分是代表了置信区间，一般置信区间的参数为95%。置信区间是用来检验自相关系数是否为0的判别标准。数据落在置信区间外代表自相关系数显著不为0，即存在显著相关关系。

```python
# 绘制日收益率数据的自相关和偏自相关图
plot_autoregression(data.d_return, 'd_return')
```

### 1.4. 信息准则

信息准则是一种模型筛选的标准。一般来说，模型复杂度越高，样本内的预测效果越好，但模型过于复杂容易造成过拟合。信息准则的思想是在模型预测能力和模型复杂度之间寻求平衡。常用的信息准则包括 AIC 和 BIC。

#### 1.4.1. AIC（赤池信息准则）

AIC的表达式为：$AIC = 2K - 2 * ln(L)$

其中，K 为模型参数的数量，L 为似然函数的最大值（似然函数是评价模型拟合度的函数，模型拟合度越高似然函数的值越大）。

使用过程中需要注意，AIC只是一种进行模型比较的相对指标，并不能提供模型优劣的绝对信息。其他条件相同时，选择AIC较小的模型，即参数数量相同时选择拟合度更好（似然函数值更大）的模型，拟合度相同时选择参数更少（更简单）的模型。

#### 1.4.2. BIC（贝叶斯信息准则）

贝叶斯信息准这是另一种模型筛选指标，其表达式为：

$BIC = K * ln(n) - 2 * ln(L)$

其中，K 为模型参数的数量，n 为样本数量，L 为似然函数的最大值。BIC 与 AIC 类似，都是在鼓励提高拟合度的同时惩罚模型复杂度，BIC 更小的模型更好；不同之处在于，BIC 考虑了样本数量，对模型参数数量的惩罚要比 AIC 更大。

### 1.5 拟合优度

拟合优度是用来评价模型对数据拟合程度好坏的指标，线性时间序列中常用决定系数（coefficient of determination）衡量模型的拟合优度。决定系数的表达式为：

$R^2 =\displaystyle 1 - \frac{SSE}{SST}= 1 - \frac{\sum (x_t-\hat x_t)^2}{\sum ({x_t-\bar x})^2}$

其中，$\bar x$ 为样本均值，$\hat x_t$ 为模型对时间序列中第 $t$ 项的预测值。$R^2$ 的取值范围在0到1之间，越接近1说明模型对数据的拟合程度越好。$R^2$ 只对方差为常数的序列才有意义，方差不恒定的情况下相当于一些时间上的噪声会比其他时间大，这时给所有的残差相同的权重是不合理的。

由于 $R^2$ 会随着模型阶数升高而变大，为了比较不同复杂度的模型，使用模型自由度对 $R^2$ 进行调整，即得到 $R^2_{adj}$ :

$R^2_{adj} = 1 - \frac{\displaystyle \frac{SSE}{n-k-1}}{\displaystyle \frac{SST}{n-1}}= 1 - \displaystyle \frac{\hat \sigma_r^2}{\hat \sigma_t^2}=1 - \displaystyle\frac{n-1}{n-k-1}(1-R^2)$ ,

其中，n 为样本数量，k 为参数个数, $\hat \sigma_r^2$ 为残差方差的估计值， $\hat \sigma_t^2$ 为样本方差的估计值。$R^2_{adj}$ 反映了对模型复杂度进行惩罚的思想，即拟合残差相同的情况下，模型越简单参数越少 $R^2_{adj}$ 越大。

## 2. 白噪声和随机游走

时间序列分类：
* 白噪声序列：纯随机序列，没有规律，不可预测，因此在分析中出现白噪声序列即可停止
* 平稳非白噪声序列：时间序列的均值和方差都是常数，可以使用AR、MA等模型进行建模
* 非平稳序列：需要首先转化为平稳序列，然后用平稳序列建模的方法进行拟合

### 2.1. 白噪声过程

#### 2.1.1. 白噪声简介

白噪声是时间序列分析中的基本模型。时间序列分析的过程就是用不同的模型去拟合已知的时间序列，直到得到的残差成为一个白噪声过程。

白噪声过程：随机过程中任意时点上的变量 $X_t$ 服从均值为0，方差为常数的独立同分布（i.i.d.）且不同时间点上随机变量的协方差为0，这样的随机过程称为白噪声过程。如果时间序列中的随机变量服从正态分布，则这样的白噪声过程称为高斯白噪声。

关于白噪声过程：
* 不同时点上的变量相关系数为0，意味着序列为纯粹随机序列，不可预测。
* 理想状态下建模的残差应该是一个白噪声过程，说明模型已经比较好的解释了原始数据中的相关性，不能解释的部分（残差）是随机波动产生的，不可预测。如果模型的残差不是白噪声，说明模型有进一步优化的空间。

```python
# 使用numpy产生一个高斯白噪声序列
wn = pd.Series(np.random.randn(500))
wn.describe()
```

```python
# 观察白噪声序列的图像
wn.plot()
```

```python
# 绘制白噪声序列的自相关和偏自相关图
plot_autoregression(wn, 'White noise')
```

从自相关和偏自相关图上可以看到，模拟的白噪声序列中，滞后项的自相关系数都在之心区间之内，即不存在自相关关系。

#### 2.1.2. 白噪声检验

时间序列分析中出现白噪声，说明序列里已经没有更多可以挖掘的信息，可以停止分析过程。因此在时间序列建模分析前，通常会先检验序列是否为白噪声，防止进行无用的分析。白噪声检验又称为随机性检验，一般使用 Ljung-Box 检验。

LB检验的原假设是时间序列是白噪声序列，即p_value小于5%的情况下，可以拒绝原序列为白噪声序列的假设。

可以使用 statsmodels 工具库中的 acorr_ljungbox() 函数对时间序列进行白噪声检验。

acorr_ljungbox()函数的官方文档：https://www.statsmodels.org/dev/generated/statsmodels.stats.diagnostic.acorr_ljungbox.html

```python
from statsmodels.stats.diagnostic import acorr_ljungbox
```

```python
# 使用 acorr_ljungbox 函数进行 LB 检验
lbvalue, pvalue = acorr_ljungbox(wn, lags=30)
```

使用 acorr_ljunbox() 函数进行白噪声检验时，所有滞后阶数小于 lags 参数的时间序列都会进行LB检验并返回相应的p_value。

```python
# 观察 LB 检验的p_value
pvalue
```

通过 pvalue 可以看到，每阶滞后项的p值均大于0.05的显著度水平，因此不能拒绝原假设，即不能否应原序列为白噪声。

```python
# 对日收益率数据进行 LB 检验
lbvalue, pvalue = acorr_ljungbox(data.d_return, lags=30)
pvalue
```

观察pvalue值发现，所有滞后项都非常接近0，因此拒绝原假设，即认为时间序列存在相关性。

### 2.2. 随机游走模型

随机游走是最简单的非平稳模型： $x_t = x_{t-1} + w_t$，其中 $w_t$ 是一个白噪声过程。

随机游走模型在金融时间序列中应用广泛，经常用于对价格路径的模拟。

```python
# 定义函数模拟一个一阶自回归过程
def generate_ts(alpha=1, steps=300):
    w = np.random.randn(steps)
    ts = [0]
    for i in range(1, len(w)):
        ts.append(alpha * ts[i-1] + w[i])
    return ts
```

```python
# 生成随机游走序列
rw = pd.Series(generate_ts(1))
rw.plot()
```

```python
# 使用 ADF 检验检测随机游走过程的平稳性
adf_test(rw)
```

```python
# 绘制随机游走序列的自相关和偏自相关图
plot_autoregression(rw, 'Random Walk', 30)
```

随机游走序列的特点的是自回归系数数值都比较大，且衰减比较慢，一阶偏自相关系数与0差异显著。

```python
# 观察一阶差分以后的相关和自相关系数
plot_autoregression(close_diff, 'Close')
```

经过一阶差分后，自相关系数相比未差分时明显减小。

## 3. 平稳时间序列模型

### 3.1. 自回归模型（Auto-Regressive Model -- AR)

自相关模型本质上是一个线性回归模型, p阶自回归模型AR(p)中，当前随机变量的值是过去p个滞后项的线性组合：

$\displaystyle x_t = a_1 x_{t-1} + \ldots + a_p x_{t-p} + w_t = \sum_{i=1}^p a_i x_{t-i} + w_t$

其中 $w_t$ 是一个白噪声过程。

这里隐含的假设是通过过去的变量值可以预测未来的变量值，即随机变量存在自相关性。

以下通过模拟一个自回归过程观察自回归模型的属性：

### 3.2. 移动平均模型（Moving Average Model -- MA)

平稳时间序列模型中，另一种比较简单常用的模型是移动平均模型。移动平均模型的构建思路是用近期的滞后项的加权平均值预测当前值，即：

$x_t = \displaystyle \sum^q_{i=1}b_ix_{t-i} + w_t$

当时间序列的均值为一个常数时，q阶移动平均模型MA(q)可以表示为：

$x_t = \mu + w_t +b_1w_{t-1}+b_2w_{t-2}+\dots + b_qw_{t-q}$

其中 $w_t$  是一个白噪声过程, $\mu$ 是时间序列的均值。

根据对移动平均模型特性的分析可以知道，移动平均模型在ACF图上是q阶截尾的，因此常用ACF图对MA模型进行定阶。

####  3.2.1. 移动平均过程模拟

```python
# 观察移动平均序列的自相关和偏自相关图
plot_autoregression(ma_series)
```

从图上可以看到，移动平均模型的自相关图截尾，偏自相关图拖尾，这一特征可以作为判定MA模型阶数的依据。

#### 3.2.2. 移动平均模型拟合

### 3.3. 自回归移动平均模型（Auto-Regressive Moving Average Model -- ARMA)

ARMA模型结合了 AR 模型和 MA 模型，在某些情况下可以以更少的模型参数实现比较好的数据拟合效果。ARMA(p,q)模型的形式是：

## 5. 异方差模型

ARIMA模型已经可以处理非平稳时间序列，但仍然假设残差时间序列的方差是恒定的，但实际的金融时间序列中存在大量的异方差现象，即去除趋势性和季节性因素后，残差的方差是随时间变化的。残差不恒定会导致参数估计结果的置信区间更窄，即低估误差或高估模型的准确性。

金融学中经常使用波动率作为衡量风险的标准，对波动率的低估也即是对风险低估。ARCH/GARCH 模型以残差的方差为建模对象，目标是提供更有效的波动率的估计值，为风险决策提供依据。

### 5.1. ARCH模型

当时间序列的方差随时间变化,并且方差变化存在相关性时，可以使用自相关过程对方差进行建模。

ARCH模型的全称是 Autoregressive Conditional Heteroskedasticity，即自回归条件异方差模型。ARCH模型假设资产收益率可以被分解为均值和波动两部分：

$r_t = \mu_t + \epsilon_t$

其中:

$\mu_t$ 表示收益率的均值，

$\epsilon_t = \sigma_t z_t $

$z_t \sim i.i.d.(0,1) $

ARCH (autoregressive conditional heteroskedasticity) 模型将波动率表示为残差滞后项 $\epsilon_t$ 平方的线性组合：

$\displaystyle \sigma_t^2 = \alpha_0 + \sum_{i=1}^p \alpha_i \epsilon_{t-i}^2$

其中 $\alpha_0 > 0, \alpha_i \geq 0$ ，模型简记为 ARCH(p)，其中p是用到的残差滞后项的数量。

注意：ARCH 模型仍然要求建模数据是剔除趋势和季节因素的，实际应用过程中，经常使用 ARCH 模型对平稳模型的残差进行建模。

ARCH 模型通常使用残差平方的偏自相关图确定阶数，ARCH 模型阶数比较高的情况下，使用 GARCH 模型通常可以构造出参数更少的模型。

### 5.2. GARCH模型

如果假设残差的方差服从ARMA（自回归移动平均模型）时，就得到GARCH (generalized autoregressive conditional heteroskedasticity) 模型。

GARCH(p,q)模型：

$\displaystyle \sigma_t^2 = \alpha_0 + \sum_{i=1}^p\alpha_i\epsilon_{t-i}^2 + \sum_{i=1}^q\beta_i\sigma_{t-i}^2$

其中 $\alpha_0 > 0, \alpha_i \geq 0, \beta_i \geq 0$ , 且 $0 < \displaystyle \sum_{i=1}^p \alpha_i + \sum_{j=1}^q \beta_j < 1$ 。最后一个条件可以保证无条件方差有限且不变。

GARCH 模型可以理解为波动率同时受到前m个残差和前n个波动率的影响。

当m取0时，GARCH 模型就退化为 ARCH 模型。

### 5.3. 异方差分析

异方差模型的建模流程：
1. 对时间序列建立线性模型，对线性模型的残差做白噪声检验
2. 如果残差通过白噪声检验，对残差进行ARCH检验（即对残差平方进行白噪声检验）
3. 异方差检验结果显著的情况下，构建异方差模型
4. 模型验证和改进

#### 5.3.2. 异方差性检验

#### 5.3.4. 模型检验

通过对残差进行标准化检验异方差模型的拟合效果。

$\epsilon_{std} = \displaystyle \frac{\epsilon_t}{\sigma_t}$  

对标准化残差进行白噪声检验可以检查均值模型的充分性，对标准化残差的平方进行白噪声检验可以检查异方差模型的充分性。

```python
# 观察标准化残差
resid = ar_fitted.resid
sigma = garch_fitted.conditional_volatility
resid_std = resid / sigma
resid_std.plot(lw=1)
```

```python
# 标准化残差的自相关和偏自相关图图
plot_autoregression(resid_std)
```

```python
# 对标准化残差进行随机性检验
lbvalue, pvalue = acorr_ljungbox(resid_std, lags=30)
pvalue
```

```python
# 对标准化残差的平方进行随机性检验
lbvalue, pvalue = acorr_ljungbox(resid_std ** 2, lags=30)
pvalue
```

从白噪声检验和自相关图上可以看到，标准化残差和标准化残差平方都为白噪声，均值模型和异方差模型都已经充分建模。

#### 5.3.5. 模型预测

使用 $\hat{\mu} \pm 1.96\hat{\sigma}_t$ 估计收益率的波动区间:

```python
# 计算收益率的均值
mu = data.d_return.mean()
mu
```

```python
# 计算历史波动率和对应的95%的收益率置信区间
data['hist_vol'] = data.d_return.rolling(window=30).std()
data['hist_vol_pos'] = mu + data.hist_vol * 1.96
data['hist_vol_neg'] = mu - data.hist_vol * 1.96
```

```python
# 使用模型预测波动率
forecasts = garch_fitted.forecast(horizon=1, start=data.index[0])
sigma_hat = np.sqrt(forecasts.variance)
sigma_hat.head()
```

```python
# 根据模型预测的波动率计算收益率95%的置信区间
data['sigma_hat'] = sigma_hat['h.1']
data['garch_pos'] = mu + data.sigma_hat * 1.96
data['garch_neg'] = mu -data.sigma_hat * 1.96
data.head()
```

```python
# 日收益率及95%的置信区间
plt.figure(figsize=(15,6))
data.d_return.plot(c='cornflowerblue', lw=1, label='daily_return', legend=True)
data.garch_neg.plot(c='orangered', ls='--', lw=1, label='garch_interval', legend=True)
data.garch_pos.plot(c='orangered', ls='--', lw=1)
data.hist_vol_neg.plot(c='g', ls='--', lw=1, label='hist_label', legend=True)
data.hist_vol_pos.plot(c='g', ls='--', lw=1)
```
